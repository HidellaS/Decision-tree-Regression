# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
print(diabetes['DESCR'])
```

```python
diabetes['feature_names']
```

```python
X = pd.DataFrame(diabetes['data'], columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
X
```

```python
y = diabetes['target']
y
```

# Split data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1011)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Check correlation

```python
X_train.corr()
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(X_train.corr(), annot=True)
```

# Create the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train, y_train)
```

# Hyper parameter tuning using grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}

params
```

```python
grid=GridSearchCV(regressor,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

# Check the performance of the model

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
selectedmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='log2',splitter='random')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
##Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(selectedmodel,filled=True)
plt.tight_layout();
```
